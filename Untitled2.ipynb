```python
from datetime import datetime

class ParkingLot:
    def __init__(self, capacity):
        # Total parking slots
        self.capacity = capacity

        # Store vehicle details
        self.slots = []

        # Counters
        self.incoming_count = 0
        self.outgoing_count = 0

    # 🚘 Park Vehicle (Incoming)
    def park_vehicle(self, vehicle_number, vehicle_type):
        if len(self.slots) < self.capacity:

            slot_number = len(self.slots) + 1
            entry_time = datetime.now()

            vehicle_data = {
                "slot_no": slot_number,
                "vehicle_no": vehicle_number,
                "vehicle_type": vehicle_type,
                "entry_time": entry_time
            }

            self.slots.append(vehicle_data)
            self.incoming_count += 1

            print(f"Vehicle {vehicle_number} parked at Slot {slot_number}")

        else:
            print("❌ Parking Full! No space available.")

    # 🚗 Remove Vehicle (Outgoing)
    def remove_vehicle(self, vehicle_number):
        for vehicle in self.slots:
            if vehicle["vehicle_no"] == vehicle_number:

                exit_time = datetime.now()
                parked_time = exit_time - vehicle["entry_time"]

                hours = parked_time.total_seconds() / 3600

                # Fee Calculation (₹20 per hour)
                fee = max(20, int(hours * 20))

                print(f"Vehicle {vehicle_number} exited.")
                print(f"Parking Time: {round(hours,2)} hours")
                print(f"Parking Fee: ₹{fee}")

                self.slots.remove(vehicle)
                self.outgoing_count += 1
                return

        print("❌ Vehicle not found.")

    # 📊 Show Parking Status
    def show_status(self):
        print("\n--- Current Parking Status ---")

        if not self.slots:
            print("Parking is empty.")
        else:
            for vehicle in self.slots:
                print(f"Slot {vehicle['slot_no']} | "
                      f"{vehicle['vehicle_type']} | "
                      f"{vehicle['vehicle_no']} | "
                      f"Entry: {vehicle['entry_time'].strftime('%H:%M:%S')}")

        print(f"\nTotal Slots: {self.capacity}")
        print(f"Occupied: {len(self.slots)}")
        print(f"Available: {self.capacity - len(self.slots)}")

    # 📈 Show Statistics
    def show_statistics(self):
        print("\n--- Parking Statistics ---")
        print(f"Total Incoming Vehicles : {self.incoming_count}")
        print(f"Total Outgoing Vehicles : {self.outgoing_count}")
```

```python
# Create Parking Lot with 3 spaces
p1 = ParkingLot(3)

while True:
    print("\n1. Park Vehicle")
    print("2. Remove Vehicle")
    print("3. Show Status")
    print("4. Show Statistics")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        v_no = input("Enter Vehicle Number: ")
        v_type = input("Enter Vehicle Type (Car/Bike): ")
        p1.park_vehicle(v_no, v_type)

    elif choice == "2":
        v_no = input("Enter Vehicle Number to Remove: ")
        p1.remove_vehicle(v_no)

    elif choice == "3":
        p1.show_status()

    elif choice == "4":
        p1.show_statistics()

    elif choice == "5":
        print("Exiting System...")
        break

    else:
        print("Invalid Choice!")
```

```python
class Todo:
  tasks = {}

  tid = 1
  def createTask(self):
    tsk = input("Enter your task :")
    dead = input("Enter deadline in (dd//mm/yyyy) :")
    prior = int(input("Enter priority :"))
    tg = input("Enter task tag (Pending):")
    self.tasks[self.tid] = {'Task':tsk,'Deadline':dead,'Priority':prior,'tag':tg}
    self.tid+=1

  def changeTaskStatus(self,n):
    status = input("Enter the status of the task :")
    if n not in self.tasks:
      print("Task ID not found !!")
      return
    else:
      self.tasks[n]['tag'] = status

  def filterTask(self,status):
    print(f"Tasks filtered according to {status}")
    for i in self.tasks:
      if self.tasks[i]['tag'] == status:
        print(f"TaskID : {i}, Task : {self.tasks[i]['Task']} ,Deadline : {self.tasks[i]['Deadline']}")

  def sortByPriority(self):
    sortTasks = sorted(self.tasks.items(),
                          key=lambda item: item[1]['Priority'],reverse=True)
    for tid, task in sortTasks[::-1]:
        print(tid, task)

  def viewTask(self):
    for i in self.tasks:
      print(f"TaskID : {i}, Task : {self.tasks[i]['Task']} ,Deadline : {self.tasks[i]['Deadline']}")


t1 = Todo()
n = int(input("Enter the number of tasks to enter :"))
for i in range(n):
  t1.createTask()

print("All the tasks are :")
t1.viewTask()

n2 = int(input("Enter the number of tasks to change the status :"))
for j in range(n2):
  z = int(input("Enter the task ID :"))
  t1.changeTaskStatus(z)

print("View tasks by tag :")
tg =input("Enter tag to sort :")
t1.filterTask(tg)

print("View Task based on priority : ")
t1.sortByPriority()
```

```python
class Store:
  def __init__(self):
    self.products = {}
    self.pid = 1

  def addProduct(self):
    prod = input("Enter your Product name :")
    stock = int(input("Enter stock number :"))
    pr = float(input("Enter its price:"))
    disc = int(input("Enter the discount :"))
    discPrice = pr - (disc/100)*pr
    self.products[self.pid] = {'Product':prod,'stock':stock,'OPrice':pr,"Discount":disc,"DiscPr":discPrice}
    self.pid+=1

  def changeDiscount(self):
    id = int(input("Enter the id of the product :"))
    if id not in self.products:
      print("Product not found !!")
      return
    disc = int(input("Enter the discount :"))
    pr = self.products[id]['OPrice']
    discPrice = pr - (disc/100)*pr
    self.products[id]['Discount'] = disc
    self.products[id]['DiscPr'] = discPrice
    print("Discount updated successfully !!")

  def viewProduct(self):
    for i in self.products:
      p = self.products[i]
      print(f"ProductID : {i}, Product : {p['Product']} ,stock : {p['stock']} ,Price : {p['OPrice']} ,Discount Price : {p['DiscPr']}")

  def makePurchase(self):
    print("All product :")
    self.viewProduct()
    cart = []
    n = int(input("Enter the no of item to purchase :"))
    total = 0
    for i in range(n):
      pd = int(input("Enter the product ID :"))
      if pd not in self.products:
        print("Invalid Product ID")
        continue
      quan = int(input("Enter the number of quantity :"))
      if quan > self.products[pd]['stock']:
        print("Quantity is greater then available stock")
        continue
      price = self.products[pd]['DiscPr']
      cost = price * quan
      self.products[pd]['stock'] -= quan
      cart.append((pd,quan,cost))
      total += cost
    print("Cart details :")
    for c in cart:
      print("ProductID :",c[0],"Quantity :",c[1],"Cost :",c[2])
    print("Total Bill :",total)

  def sortByPrice(self):
    s = sorted(self.products.items(), key=lambda x: x[1]['DiscPr'])
    for pid,p in s:
      print(pid,p)


s1 = Store()

n = int(input("Enter the number of products to add :"))
for i in range(n):
  s1.addProduct()

print("All the products are :")
s1.viewProduct()

n2 = int(input("Enter number of products to change discount :"))
for j in range(n2):
  s1.changeDiscount()

print("Products after discount update :")
s1.viewProduct()

print("Make Purchase :")
s1.makePurchase()

print("Products sorted by price :")
s1.sortByPrice()
```

```python
class Theatre:

    def __init__(self):
        self.movieName = "MS DHONI"
        self.seats = {i:"Available" for i in range(1,21)}
        self.price = 100

    def checkSeats(self):
        print("Seat Status for", self.movieName)
        for s in self.seats:
            print("Seat", s, ":", self.seats[s])

    def selectSeats(self):
        self.checkSeats()
        n = int(input("Enter number of seats you want to select: "))
        selectedSeats = []
        for i in range(n):
            seat = int(input("Enter seat number: "))
            if seat not in self.seats:
                print("Invalid seat number")
                continue
            if self.seats[seat] == "Booked":
                print("Seat already booked")
                continue
            selectedSeats.append(seat)
        return selectedSeats

    def bookTickets(self):
        seats = self.selectSeats()
        if len(seats) == 0:
            print("No seats selected")
            return
        for s in seats:
            self.seats[s] = "Booked"
        self.printConfirmation(seats)

    def printConfirmation(self, seats):
        print("Ticket Information :")
        print("Movie:", self.movieName)
        print("Seats Booked:", seats)
        print("Total Tickets:", len(seats))
        print("Total Price:", len(seats)*self.price)
        print("Booking Successful!")


t1 = Theatre()

while True:
    print("\n1. Check Seats")
    print("2. Book Tickets")
    print("3. Exit")
    ch = int(input("Enter choice: "))
    if ch == 1:
        t1.checkSeats()
    elif ch == 2:
        t1.bookTickets()
    elif ch == 3:
        print("Thank you!")
        break
    else:
        print("Invalid Choice")
```

```python
class College:
    def __init__(self):
        self.Sem1 = {i:0 for i in range(1,6)}
        self.semCredit = 20
        self.Sem2 = {i:0 for i in range(1,6)}
        self.CGPA = {i:'NA' for i in range(1,6)}
        self.Percentage = {i:0 for i in range(1,6)}

    def checkSem1Marks(self):
        print("Python")
        for i in self.Sem1:
            print(f"Roll {i} SGPA {self.Sem1[i]}")

    def checkSem2Marks(self):
        print("JAVA")
        for i in self.Sem2:
            print(f"Roll {i} SGPA {self.Sem2[i]}")

    def allotCGPA(self):
        print("Allot CGPA :")
        for i in range(1,6):
            self.Sem1[i] = float(input(f"Enter Sem1 SGPA of {i} :"))
            self.Sem2[i] = float(input(f"Enter Sem2 SGPA of {i} :"))
        self.calculateCGPA()

    def calculateCGPA(self):
        for i in range(1,6):
            avg = round((self.Sem1[i] + self.Sem2[i]) / 2, 3)
            self.CGPA[i] = avg
            self.Percentage[i] = avg*9.5

    def showCGPA(self):
      print("CGPA")
      for i in self.CGPA:
          print(f"Roll {i} CGPA {self.CGPA[i]}")

    def showPercentage(self):
      print("Percentage")
      for i in self.Percentage:
          print(f"Roll {i} Percentage {self.Percentage[i]}")

t1 = College()
while True:
    print("1. Check Sem1 SGPA")
    print("2. Check sem2 SGPA")
    print("3. Allot SGPA")
    print("4. Show CGPA")
    print("5. Show Percentage")
    print("6. Exit")
    ch = int(input("Enter choice: "))
    if ch == 1:
        t1.checkSem1Marks()
    elif ch == 2:
        t1.checkSem2Marks()
    elif ch == 3:
        t1.allotCGPA()
    elif ch == 4:
        t1.showCGPA()
    elif ch == 5:
        t1.showPercentage()
    elif ch == 6:
        print("Thank you!")
        break
    else:
        print("Invalid Choice")
```

```python
class VaultManager:
    def __init__(self):
        self.uid = 0
        self.data_store = {}

    def addEntry(self):
        website_name = input("Enter website name: ")
        user_mail = input("Enter username or email: ")
        pass_word = input("Enter password: ")

        self.data_store[self.uid] = {
            "Website": website_name,
            "UserMail": user_mail,
            "PassWord": pass_word
        }

        print("Entry added successfully with ID:", self.uid)
        self.uid += 1

    def addMultiple(self):
        count = int(input("Enter number of entries: "))
        for i in range(count):
            self.addEntry()

    def viewAll(self):
        if len(self.data_store) == 0:
            print("No entries available")
        else:
            print("Saved Password Records:")
            for key in self.data_store:
                info = self.data_store[key]
                print(f"ID:{key} Website:{info['Website']} Username/Email:{info['UserMail']} Password:{info['PassWord']}")

    def findPassword(self):
        key = int(input("Enter ID to view password: "))
        if key in self.data_store:
            info = self.data_store[key]
            print("Website:", info["Website"])
            print("Username or Email:", info["UserMail"])
            print("Password:", info["PassWord"])
        else:
            print("Invalid ID")

    def updateEntry(self):
        key = int(input("Enter ID to update password: "))
        if key in self.data_store:
            new_password = input("Enter new password: ")
            self.data_store[key]["PassWord"] = new_password
            print("Password updated successfully")
        else:
            print("Invalid ID")


vm = VaultManager()

while True:
    print("\n1. Add Entry")
    print("2. Add Multiple Entries")
    print("3. View All Entries")
    print("4. Get Password")
    print("5. Update Password")
    print("6. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        vm.addEntry()
    elif choice == 2:
        vm.addMultiple()
    elif choice == 3:
        vm.viewAll()
    elif choice == 4:
        vm.findPassword()
    elif choice == 5:
        vm.updateEntry()
    elif choice == 6:
        print("Thank you")
        break
    else:
        print("Invalid choice")
```

```python
class SecureStore:
    def __init__(self):
        self.index = 0
        self.records = {}

    def addOne(self):
        website = input("Enter website name: ")
        user_id = input("Enter username or email: ")
        pass_key = input("Enter password: ")

        self.records[self.index] = {
            "Website": website,
            "UserID": user_id,
            "PassKey": pass_key
        }

        print("Password stored successfully with ID:", self.index)
        self.index += 1

    def addMany(self):
        total = int(input("How many passwords you want to add: "))
        for i in range(total):
            self.addOne()

    def showAll(self):
        if len(self.records) == 0:
            print("No password saved")
        else:
            print("Saved Password List")
            for key in self.records:
                data = self.records[key]
                print(f"ID:{key} Website:{data['Website']} Username/Email:{data['UserID']} Password:{data['PassKey']}")

    def searchPassword(self):
        key_id = int(input("Enter ID to view password: "))
        if key_id in self.records:
            data = self.records[key_id]
            print("Website:", data["Website"])
            print("Username/Email:", data["UserID"])
            print("Password:", data["PassKey"])
        else:
            print("Invalid ID")

    def updatePassword(self):
        key_id = int(input("Enter ID to change password: "))
        if key_id in self.records:
            new_pass = input("Enter new password: ")
            self.records[key_id]["PassKey"] = new_pass
            print("Password changed successfully")
        else:
            print("ID not found")


store = SecureStore()

while True:
    print("\n1. Add One Password")
    print("2. Add Multiple Passwords")
    print("3. Display All Passwords")
    print("4. Find Password")
    print("5. Update Password")
    print("6. Exit")

    option = int(input("Enter your option: "))

    if option == 1:
        store.addOne()
    elif option == 2:
        store.addMany()
    elif option == 3:
        store.showAll()
    elif option == 4:
        store.searchPassword()
    elif option == 5:
        store.updatePassword()
    elif option == 6:
        print("Program closed")
        break
    else:
        print("Wrong choice")
```
